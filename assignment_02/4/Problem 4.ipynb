```python
from gmm import GaussianMixtureModel
import cv2
import matplotlib.pyplot as plt
```

```python
peppers_color = cv2.imread('peppers_color.tif')
data = peppers_color.reshape(-1, 3)
model = GaussianMixtureModel(num_of_clusters=5, num_of_iterations=100, reshape_to=(peppers_color.shape[0], peppers_color.shape[1]))
model.train(data)
clusters = model.test(data)
```

```python
plt.figure()
plt.imshow(clusters)
plt.savefig('segmented_img.jpeg', dpi=300)
plt.show()
```
